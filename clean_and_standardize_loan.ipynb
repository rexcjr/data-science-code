```python
import pandas as pd
import numpy as np

df = pd.read_csv('sales.csv') #load csv file to df
df.describe(include="all") # show statistical summary
```

```python
df.info()
```

```python
df_nodups = df.drop_duplicates() # make sure there are no duplicates in the dataset
df_nodups
```

```python
df_nodups.info()
```

```python
int_price = {"13,000":"13000"} # set value to be replaced 
df1 = df_nodups.replace({"Price":int_price}) #replace value in order to convert str to int/float
df1.Price = pd.to_numeric(df1.Price) #convert data type to numeric

df1.info()
```

```python
#remove outliers first before standardizing Price column
m = np.mean(df1.Price)
s = np.std(df1.Price)

print("Mean: ",m)
print("Standard Deviation: ",s)

ll = max(0,m - 3*s)
ul = m + 3*s

print("Lower Limit: ",ll)
print("Upper Limit: ",ul)

df1_cleaned = df1[df1.Price >= ll]
df1_cleaned = df1_cleaned[df1_cleaned.Price <= ul] 

df1_cleaned
```

```python
# Standardize Price Column

df1_cleaned.insert(3, "standardized_price", 0) # insert column 
df2 = df1_cleaned
```

```python
#df2['standardized_price'] = df2.loc[df2['Price']].apply(lambda x: (x - np.mean(x) ) / (np.std(x) ))
#df2

df2.standardized_price = ((df2.Price - np.mean(df2.Price)) / (np.std(df2.Price)) )
df2
```

```python
df2.to_csv('clean_sales.csv', index = False, header = True)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('Handling_Missing_Values.csv')
data
```

```python
data.describe(include="all")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop([9,0],inplace=True) #removed row index 9 since it has 3 blank values
data
```

```python
# Data Imputation

# we can create a new category since some may have chosen not to provide their sex
data['Sex'].fillna(value="Undetermined", inplace = True)

# used the mean of Income since the range is just 1875 (Max-Min) and no correlation with other data
data['Income'].fillna(value=round(np.mean(data.Income),0), inplace = True)

# also used mean for children and rounded off value
data['Children'].fillna(value=round(np.mean(data.Children),0), inplace = True)

data.isnull().sum()
```

```python
data
```

```python
data.to_csv('clean_handling_missing_values.csv', index = False, header = True)
```

```python
import pandas as pd
import numpy as np

ds = pd.read_csv('dataset.csv')
ds
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds.isnull().sum()
```

```python
ds.drop([12,0],inplace=True) #removed row index 12 since it has many blank values
ds
```

```python
# we can create a new category since some may have chosen not to provide their sex
ds['gender'].fillna(value="Undetermined", inplace = True)
ds['net_income'].fillna(value=round(np.mean(ds.net_income),0), inplace = True) #mean net income
ds['dependents'].fillna(value=round(np.mean(ds.dependents),0), inplace = True) #mean dependents

ds.isnull().sum()
```

```python
x = np.mean(ds.loan_amount)
y = np.std(ds.loan_amount)

print("Mean: ",x)
print("Standard Deviation: ",y)

ll2 = max(0,x - 3*y)
ul2 = x + 3*y

print("Lower Limit: ",ll2)
print("Upper Limit: ",ul2)

ds1 = ds[ds.loan_amount >=ll2]
ds2 = ds1[ds1.loan_amount <=ul2] 

ds2
```

```python
ds2.insert(6, "standardized_loan", 0) # insert column 

ds2.standardized_loan = (ds2.loan_amount - np.mean(ds2.loan_amount))/np.std(ds2.loan_amount) #standardize loan amount
ds2
```

```python
ds2.to_csv('cleaned_dataset.csv', index = False, header = True)
```
