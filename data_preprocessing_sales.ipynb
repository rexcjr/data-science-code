```python
# initial characterization of the entire dataframe

import pandas as pd

df = pd.read_csv("sales.csv") #load csv file to df
df.describe(include="all") # show statistical summary
```

```python
# Remove Duplicate row values

df_nodups = df.drop_duplicates() 
df_nodups
```

```python
# Express Price in Thousands

df_nodups.info() # check column data types
```

```python
int_price = {"13,000":"13000"} # set value to be replaced 
df1 = df_nodups.replace({"Price":int_price}) #replace value in order to convert str to int/float 

# Needed to be int/float data type in order to express Price column in Thousands
df1.Price = pd.to_numeric(df1.Price) / 1000 
df1
```

```python
#save updated sales.csv file
df1.to_csv(r"C:\Users\carolirs\carolino_sales.csv", index = False, header=True) 
```

```python
# Select name, city, state and country data of product 3 bought by customers from United States 
# and paid thru Visa or AMEX. Store the selected data in empty.csv file.

# filter specific values per applicable column
x = df1[(df1.Product == "Product3") & (df1.Country == "United States") & 
       ((df1.Payment_Type == "Visa") | (df1.Payment_Type == "Amex"))]

# filter needed columns
x1 = x[['Name', 'City', 'State', 'Country']] 
x1 #print to double check if filtered correctly
```

```python
# save to empty.csv file
x1.to_csv('empty.csv', index = False, header=True)
```

```python
# Select Name, Price, City, State, and Country data of Product2 bought by customers from New York 
# and paid thru Mastercard. Store the selected data in empty1.csv file.

# filter specific values per applicable column
y = df1[(df1.Product == "Product2") & (df1.City == "New York") & 
       (df1.Payment_Type == "Mastercard")]

# filter needed columns
y1 = y[['Name', 'Price', 'City', 'State', 'Country']] 
y1 #print 
```

```python
# save to empty1.csv file
y1.to_csv("empty1.csv", index = False, header=True)
```
