```python
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
%matplotlib notebook

import matplotlib as mpl
import matplotlib.pyplot as plt

df = df.T
df.head()
```

```python
yLine = 41111
y_val = []
y_error_val = []

for i in df.columns :
    y_error_val.append(1.96 * np.std(df[i]) / (len(df[i]) ** 0.5) )
    y_val.append(np.mean(df[i]) )

plt.xticks(df.columns)
bars = plt.bar(df.columns, y_val)

for i in range(0,len(y_val) ) :
    x, y, z = plt.errorbar(df.columns[i], y_val[i], y_error_val[i], 
                         color = 'black', capsize = 10, linestyle = None)

    if yLine >= y[1].get_xydata()[0,1] :
        bars[i].set_color('darkblue')
    elif yLine < y[1].get_xydata()[0,1] and yLine>y_val[i] :
        bars[i].set_color('lightblue')
    elif yLine > y[0].get_xydata()[0,1] and yLine<y_val[i] :
        bars[i].set_color('#fedcb8')
    else:
        bars[i].set_color('maroon')
        
```
